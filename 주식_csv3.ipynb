```python
#삼성전자
import pandas as pd 
import requests 

html = requests.get('https://finance.naver.com/item/sise_day.nhn?code=005930&page=1') 

table = pd.read_html(html.text)
```

```python
for i in range(50):
    html = requests.get('https://finance.naver.com/item/sise_day.nhn?code=005930&page={}'.format(i + 1))
    table = pd.read_html(html.text)
    if i == 0:
        data = table[0].dropna().reset_index(drop = True)
    else:
        data1 = table[0].dropna().reset_index(drop = True)
        data = pd.concat([data, data1])

samsung_data = data.reset_index(drop = True)
samsung_data
```

```python
#csv 파일 전환
samsung_data.to_csv('samsung.csv' ,encoding='cp949')
```

```python
#네이버
import pandas as pd 
import requests 
```

```python
idx_col= '종목명'

xlsx = pd.read_excel('JongMokCodeTable_20140616_Uodate02.xlsx', index_col = idx_col)
xlsx[['종목코드']]
```

```python
x = input()
y = xlsx[['종목코드']].loc[x]
z = y['종목코드'].head(1).iloc[0].replace("'","")

html = requests.get('https://finance.naver.com/item/sise_day.nhn?code={}&page=1'.format(z))

table = pd.read_html(html.text)

for i in range(30):
    html = requests.get('https://finance.naver.com/item/sise_day.nhn?code={}&page={}'.format(z, i + 1))
    table = pd.read_html(html.text)
    if i == 0:
        data = table[0].dropna().reset_index(drop = True)
    else:
        data1 = table[0].dropna().reset_index(drop = True)
        data = pd.concat([data, data1])

naver_data = data.reset_index(drop = True)
naver_data
```

```python

import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker 
from mpl_finance import candlestick2_ohlc 
from matplotlib import font_manager, rc
```

```python
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

fig, ax = plt.subplots() 

test_data = naver_data[::-1]
xdate = test_data.날짜.astype('str')
for i in range(len(xdate)):
    xdate[i] = xdate[i][2:] 

ax.plot(xdate, test_data['종가'], label="종가",linewidth=0.7,color='k')
ax.plot(xdate, test_data['종가'].rolling(window=5).mean(), label="평균5일",linewidth=0.7)
ax.plot(xdate, test_data['종가'].rolling(window=20).mean(), label="평균20일",linewidth=0.7)
ax.plot(xdate, test_data['종가'].rolling(window=60).mean(), label="평균60일",linewidth=0.7)
#ax.plot(xdate, naver_data['종가'].rolling(window=120).mean(), label="평균120일",linewidth=0.7)
fig.suptitle("개미들의 주식전쟁",fontsize=30)
ax.set_xlabel("날짜")
ax.set_ylabel("주가(원)")
ax.xaxis.set_major_locator(ticker.MaxNLocator(25))
fig.set_size_inches(18.5, 10.5)

ax.legend(loc=1)
plt.xticks(rotation = 45)

plt.grid()
plt.show()
```
